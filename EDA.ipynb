```python
import pandas as pd
import numpy as np
```

```python
# Se importa el archivo csv con los datos de los accidentes.

db = pd.read_csv ('AccidentesAviones.csv')
db
```

```python
db.rename(columns= {'Unnamed: 0': 'ID', 'OperadOR': 'Operador', 'PASAJEROS A BORDO': 'passenger_aboard', 'ground':'ground_fatalities'}, inplace= True)
```

```python
db.info()
```

```python
# Remplazamos los valores '?' a NaN

db = db.replace('?', np.NaN)
```

```python
db.columns
```

```python
# Convertimos a int las siguientes columnas:

columns = ['all_aboard', 'passenger_aboard', 'crew_aboard', 'cantidad de fallecidos', 'passenger_fatalities', 'crew_fatalities', 'ground_fatalities']
for column in columns:
    db[column]= pd.to_numeric(db[column]).convert_dtypes(int)

db.info()

db
```
